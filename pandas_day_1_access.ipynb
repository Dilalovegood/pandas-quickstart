```python
import pandas as pd

data = {
    "Day": ["Monday", "Monday", "Tuesday", "Tuesday", "Wednesday", "Wednesday", "Thursday", "Thursday", "Friday", "Friday", "Saturday", "Saturday", "Sunday", "Sunday"],
    "Coffee Type": ["Espresso", "Latte", "Espresso", "Latte", "Espresso", "Latte", "Espresso", "Latte", "Espresso", "Latte", "Espresso", "Latte", "Espresso", "Latte"],
    "Units Sold": [25, 15, 30, 20, 35, 25, 40, 30, 45, 35, 45, 35, 45, 35]
}
coffee = pd.DataFrame(data)
```

```python
coffee = coffee.reset_index(drop=True)
```
